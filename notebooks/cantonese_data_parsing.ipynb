```python
import pandas as pd
import json
from collections import defaultdict
```

```python
loc ='../../pycantonese/src/pycantonese/data/rime_cantonese/chars_to_jyutping.json'
with open(loc,'rb') as f:
    loaded=json.load(f)
```

```python

reversed_mapping = defaultdict(list)
# reverse structure
for k,v in loaded.items():
    reversed_mapping[v].append(k)
```

```python
canto_df = []
for k,v in reversed_mapping.items():
    num_parts = sum(c.isdigit() for c in k)
    if num_parts ==1:
        tone = list(k)[-1]
        phoneme = ''.join(list(k)[:-1])

        canto_df.append({'label':k,
                        'phonenme':phoneme,
                        'tone':tone,
                        'chareters':v,
                        'charecter_count':len(v)})
    else:


        canto_df.append({'label':k,
                        'phonenme':None,
                        'tone':None,
                        'chareters':v,
                        'charecter_count':len(v)})

```

```python
df_canto = pd.DataFrame(canto_df)
# df_canto[~df_canto.phonenme.isna()]
```
